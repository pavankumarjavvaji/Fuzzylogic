```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as cntrl
```

```python
ATTENDANCE = 'Attendance'
PERFORMANCE = 'Performance'
INTERNAL_MARKS = 'Internal_Marks'
EXTERNAL_MARKS = 'External_Marks'
POOR = 'Poor'
AVERAGE = 'Average'
GOOD = 'Good'
V_GOOD = 'V.Good'
EXCELLENT = 'Excellent'
low_parameter = [0,10,40,50]
average_parameter = [30,40,50,60]
good_parameter = [40,50,60,70]
v_good_parameter = [50,60,70,80]
excellent_parameter = [65,80,100,100]
```

 Create the three fuzzy variables - three inputs, one output
 uzzy logic is a mathematical framework that allows for dealing with uncertainty and imprecision in decision-making

```python
intrn_marks =cntrl.Antecedent(np.arange(0,105,5),INTERNAL_MARKS)
attendance = cntrl.Antecedent(np.arange(0,105,5), ATTENDANCE)
extrn_marks = cntrl.Antecedent(np.arange(0,105,5), EXTERNAL_MARKS)
performance = cntrl.Consequent(np.arange(0,105,5), PERFORMANCE)
```

#Trapezoidal membership function generator.
trapmf(x, abcd) x is one d array abcd is ond d array

```python
intrn_marks[POOR] = fuzz.trapmf(intrn_marks.universe, low_parameter)
intrn_marks[AVERAGE] = fuzz.trapmf(intrn_marks.universe, average_parameter)
intrn_marks[GOOD] = fuzz.trapmf(intrn_marks.universe, good_parameter)
intrn_marks[V_GOOD] = fuzz.trapmf(intrn_marks.universe, v_good_parameter)
intrn_marks[EXCELLENT] = fuzz.trapmf(intrn_marks.universe, excellent_parameter)
```

```python
attendance[POOR] = fuzz.trapmf(attendance.universe, low_parameter)
attendance[AVERAGE] = fuzz.trapmf(attendance.universe,  average_parameter)
attendance[GOOD] = fuzz.trapmf(attendance.universe, good_parameter)
attendance[V_GOOD] = fuzz.trapmf(attendance.universe, v_good_parameter)
attendance[EXCELLENT] = fuzz.trapmf(attendance.universe,  excellent_parameter)
```

```python
extrn_marks[POOR] = fuzz.trapmf(extrn_marks.universe, low_parameter)
extrn_marks[AVERAGE] = fuzz.trapmf(extrn_marks.universe, average_parameter)
extrn_marks[GOOD] = fuzz.trapmf(extrn_marks.universe, good_parameter)
extrn_marks[V_GOOD] = fuzz.trapmf(extrn_marks.universe, v_good_parameter)
extrn_marks[EXCELLENT] = fuzz.trapmf(extrn_marks.universe, excellent_parameter)

performance[POOR] = fuzz.trapmf(performance.universe, low_parameter)
performance[AVERAGE] = fuzz.trapmf(performance.universe, average_parameter)
performance[GOOD] = fuzz.trapmf(performance.universe, good_parameter)
performance[V_GOOD] = fuzz.trapmf(performance.universe, v_good_parameter)
performance[EXCELLENT] = fuzz.trapmf(performance.universe, excellent_parameter)
```

```python
rule1 = cntrl.Rule(attendance[POOR] & extrn_marks[POOR] & intrn_marks[POOR], performance[POOR])
rule2 = cntrl.Rule(attendance[POOR] & extrn_marks[AVERAGE] & intrn_marks[POOR], performance[POOR])
rule3 = cntrl.Rule(attendance[POOR] & extrn_marks[GOOD] & intrn_marks[POOR], performance[AVERAGE])
rule4 = cntrl.Rule(attendance[POOR] & extrn_marks[V_GOOD] & intrn_marks[POOR], performance[AVERAGE])
rule5 = cntrl.Rule(attendance[POOR] & extrn_marks[GOOD] & intrn_marks[V_GOOD], performance[GOOD])
rule6 = cntrl.Rule(attendance[POOR] & extrn_marks[POOR] & intrn_marks[AVERAGE], performance[POOR])
rule7 = cntrl.Rule(attendance[POOR] & extrn_marks[AVERAGE] & intrn_marks[AVERAGE], performance[AVERAGE])
rule8 = cntrl.Rule(attendance[POOR] & extrn_marks[GOOD] & intrn_marks[AVERAGE], performance[AVERAGE])
rule9 = cntrl.Rule((attendance[POOR] & extrn_marks[GOOD] & intrn_marks[GOOD]), performance[GOOD])
rule10 = cntrl.Rule(attendance[POOR] & extrn_marks[EXCELLENT] & intrn_marks[GOOD], performance[V_GOOD])
rule11 = cntrl.Rule(attendance[AVERAGE] & extrn_marks[AVERAGE] & intrn_marks[GOOD], performance[AVERAGE])
rule12 = cntrl.Rule(attendance[AVERAGE] & extrn_marks[GOOD] & intrn_marks[GOOD], performance[GOOD])
rule13 = cntrl.Rule(attendance[AVERAGE] & extrn_marks[V_GOOD] & intrn_marks[GOOD], performance[GOOD])
rule14 = cntrl.Rule(attendance[AVERAGE] & extrn_marks[V_GOOD] & intrn_marks[V_GOOD], performance[V_GOOD])
rule15 = cntrl.Rule(attendance[AVERAGE] & extrn_marks[AVERAGE] & intrn_marks[EXCELLENT], performance[GOOD])
rule16 = cntrl.Rule(attendance[AVERAGE] & extrn_marks[AVERAGE] & intrn_marks[AVERAGE], performance[AVERAGE])
rule17 = cntrl.Rule(attendance[AVERAGE] & extrn_marks[POOR] & intrn_marks[POOR], performance[POOR])
rule18 = cntrl.Rule(attendance[AVERAGE] & extrn_marks[POOR] & intrn_marks[GOOD], performance[AVERAGE])
rule19 = cntrl.Rule(attendance[GOOD] & extrn_marks[AVERAGE] & intrn_marks[AVERAGE], performance[AVERAGE])
rule20 = cntrl.Rule(attendance[GOOD] & extrn_marks[EXCELLENT] & intrn_marks[EXCELLENT], performance[V_GOOD])
rule21 = cntrl.Rule(attendance[GOOD] & extrn_marks[GOOD] & intrn_marks[AVERAGE], performance[GOOD])
rule22 = cntrl.Rule(attendance[GOOD] & extrn_marks[POOR] & intrn_marks[POOR], performance[POOR])
rule23 = cntrl.Rule(attendance[V_GOOD] & extrn_marks[EXCELLENT] & intrn_marks[V_GOOD], performance[V_GOOD])
rule24 = cntrl.Rule(attendance[V_GOOD] & extrn_marks[V_GOOD] & intrn_marks[V_GOOD], performance[V_GOOD])
rule25 = cntrl.Rule(attendance[V_GOOD] & extrn_marks[POOR] & intrn_marks[POOR], performance[POOR])
rule26 = cntrl.Rule(attendance[V_GOOD] & extrn_marks[GOOD] & intrn_marks[V_GOOD], performance[V_GOOD])
rule27 = cntrl.Rule(attendance[V_GOOD] & extrn_marks[EXCELLENT] & intrn_marks[EXCELLENT], performance[EXCELLENT])
rule28 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[EXCELLENT] & intrn_marks[V_GOOD], performance[V_GOOD])
rule29 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[AVERAGE] & intrn_marks[AVERAGE], performance[V_GOOD])
rule30 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[AVERAGE] & intrn_marks[V_GOOD], performance[GOOD])
rule31 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[AVERAGE] & intrn_marks[GOOD], performance[GOOD])
rule32 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[POOR] & intrn_marks[POOR], performance[POOR])
rule33 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[AVERAGE] & intrn_marks[POOR], performance[AVERAGE])
rule34 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[POOR] & intrn_marks[AVERAGE], performance[POOR])
rule35 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[GOOD] & intrn_marks[POOR], performance[GOOD])
rule36 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[POOR] & intrn_marks[GOOD], performance[AVERAGE])
rule37 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[V_GOOD] & intrn_marks[POOR], performance[V_GOOD])
rule38 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[POOR] & intrn_marks[V_GOOD], performance[AVERAGE])
rule39 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[POOR] & intrn_marks[EXCELLENT], performance[GOOD])
rule40 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[AVERAGE] & intrn_marks[EXCELLENT], performance[V_GOOD])
rule41 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[GOOD] & intrn_marks[EXCELLENT], performance[V_GOOD])
rule42 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[V_GOOD] & intrn_marks[EXCELLENT], performance[V_GOOD])
rule43 = cntrl.Rule(attendance[EXCELLENT] & extrn_marks[EXCELLENT] & intrn_marks[EXCELLENT], performance[EXCELLENT])
```

```python
rule_list = [rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12,
             rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, 
             rule22, rule23, rule24, rule25, rule26, rule27, rule28, rule29, rule30, 
             rule31, rule32, rule33, rule34, rule35, rule36, rule37, rule38, rule39, rule40, rule41, rule42, rule43]
```

```python
performance_ctrl = cntrl.ControlSystem(rule_list)
perf_analysis = cntrl.ControlSystemSimulation(performance_ctrl)
perf_analysis.input[ATTENDANCE] = 79
perf_analysis.input[EXTERNAL_MARKS] = 69 #example
perf_analysis.input[INTERNAL_MARKS] = 40

perf_analysis.compute()
performan(perf_analysis.output[PERFORMANCE])
print(str(perf_analysis.output[PERFORMANCE]))
```

```python
def performan(perf):
    
    if(perf<50):
        print('poor')
    elif(perf>=50 and perf<65):
        print('Average')
    elif(perf>=65 and perf<75):
        print('good')
    elif(perf>=75 and perf<85):
        print('very good')
    else:
        print('excellent')
    
```
